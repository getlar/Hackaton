```python
import numpy as np
from waitersimporter import get_waiters, display_floor
waiters_array = get_waiters("waiters.pyt")

display_floor(waiters_array[0, :, :])
display_floor(waiters_array[1, :, :])
```

```python
collision_count = 0

for t in range(waiters_array.shape[0]):
    floor_snapshot = waiters_array[t, :, :]
    
    waiter_positions = np.argwhere(np.isin(floor_snapshot, [1, 2, 3, 4]))
    for i in range(len(waiter_positions)):
        for j in range(i + 1, len(waiter_positions)):
            pos1, pos2 = waiter_positions[i], waiter_positions[j]

            if pos1[0] == pos2[0]:
                dist = abs(pos1[1] - pos2[1])
                if dist <= 4:
                    if (
                        floor_snapshot[pos1[0], pos1[1]] == 1 and 
                        floor_snapshot[pos2[0], pos2[1]] == 2
                    ):
                        path_clear = all(
                            floor_snapshot[pos1[0], y] not in [5, 1, 2, 3, 4]
                            for y in range(pos1[1] + 1, pos2[1])
                        )
                        if path_clear:
                            collision_count += 1
                            print(f"Collision detected at time {t} between {pos1} and {pos2}")

            elif pos1[1] == pos2[1]:
                dist = abs(pos1[0] - pos2[0])
                if dist <= 4:
                    if (
                        floor_snapshot[pos1[0], pos1[1]] == 3 and
                        floor_snapshot[pos2[0], pos2[1]] == 4
                    ):
                        bad = False)
                        for x in range(pos2[0] + 1, pos1[0]):
                            print(f"Checking {x, pos1[1]}")
                            print(floor_snapshot[x, pos1[1]])
                            if floor_snapshot[x, pos1[1]] in [5, 1, 2, 3, 4]:
                                bad = True
                                break
                        if not bad:
                            collision_count += 1
                            print(f"Collision detected at time {t} between {pos1} and {pos2}")
                            
    print(f"Collision count at time {t}: {collision_count}")

collision_count
```
