```python
from waitersimporter import get_waiters, display_floor
import numpy as np

collision_counter = 0
floor_size = 20

def count_row_collisions(row) -> int:
    collisions_in_row = 0
    for i in range(floor_size):
        if row[i] != 1:
            continue
        else:
            for j in range(i+1, min(floor_size, i+5)):
                if row[j] == 2:
                    collisions_in_row += 1
                    break
                elif row[j] in (1, 3, 4, 5):
                    break
    return collisions_in_row

def count_col_collision(col):
    collisions_in_col = 0
    for i in range(floor_size):
        if col[i] != 3:
            continue
        else:
            for j in range(i+1, min(floor_size, i+5)):
                if col[j] == 4:
                    collisions_in_col += 1
                    break
                elif col[j] in (1, 2, 3, 5):
                    break
    return collisions_in_col


for timeslot in range(180):

    floor = waiters_array[timeslot, :, :]
    for idx_ in range(floor_size):
        
        collision_counter += count_row_collisions(floor[idx_, :])
        collision_counter += count_col_collision(floor[:, idx_])

            
print(f"Az adott idszakban Sanyi {collision_counter} alkalommal rikcsolt.")
```
