```python
from fastbook import *
import pathlib

plt = platform.system()
if plt == 'Linux':
    pathlib.WindowsPath = pathlib.PosixPath

# Ha még nem dolgoztál Deep Learning megoldásokkal, az alábbiak segíthetnek

learn_inf = load_learner('export.pkl')
kepek_eleresi_utvonalai = get_image_files('test_images/test_images/')
test_dl = learn_inf.dls.test_dl(kepek_eleresi_utvonalai)
```

```python
from IPython.display import display

preds, _ = learn_inf.get_preds(dl=test_dl)

class_1_indices = (preds.argmax(dim=1) == 1).nonzero().flatten()

total_sum = 0
for idx in class_1_indices:
    img_path = kepek_eleresi_utvonalai[idx]
    img = PILImage.create(img_path)
    filename = img_path.name
    print(f"Filename: {filename}")
    
    numbers = [int(s) for s in re.findall(r'\d+', filename)]
    print(numbers)
    total_sum += int(numbers[0])

print(f"\nTotal sum of numbers in all filenames: {total_sum}")
```
